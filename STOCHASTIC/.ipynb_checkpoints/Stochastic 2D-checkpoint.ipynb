```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from scipy.sparse import diags, eye, kron
from scipy.sparse.linalg import eigsh

# =========================
# ONE SIMPLE TOGGLE
# =========================
SHOW_W = False   # True: show W_rescaled contour, False: hide it

# ----------------------------------------------------------------------
# 1) symbolic definitions
# ----------------------------------------------------------------------
phi1, phi2 = sp.symbols('phi1 phi2', real=True)
alpha, beta, delta, gamma, xi = sp.symbols('alpha beta delta gamma xi', real=True)

Omega  = 1 + alpha + sp.sqrt(beta)
OmegaP = 1 + delta + sp.sqrt(gamma)

v = (
    (sp.pi**2 / (3 * Omega)) * (
        4*alpha*phi1
        - 2*(beta/Omega)*phi1**2
        + (4/Omega**3)*phi1**4
    )
    +
    (sp.pi**2 / (3 * OmegaP)) * (
        4*delta*phi2
        - 2*(gamma/OmegaP)*phi2**2
        + (4/OmegaP**3)*phi2**4
    )
    +
    (sp.pi**2 / (3 * Omega * OmegaP)) * xi * phi1**2 * phi2**2
)

v_phi1  = sp.diff(v, phi1)
v_phi2  = sp.diff(v, phi2)
v_phi11 = sp.diff(v_phi1, phi1)
v_phi22 = sp.diff(v_phi2, phi2)

v_func   = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v, 'numpy')
v1_func  = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi1, 'numpy')
v2_func  = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi2, 'numpy')
v11_func = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi11, 'numpy')
v22_func = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi22, 'numpy')

# ----------------------------------------------------------------------
# 2) parameters
# ----------------------------------------------------------------------
alpha_val = 0.2
beta_val  = 2.2

delta_val = 0.3
gamma_val = 2.0

xi_val = 0.2
lambda_val = 0.05

Omega_val  = 1 + alpha_val + np.sqrt(beta_val)
OmegaP_val = 1 + delta_val + np.sqrt(gamma_val)

# Since we want Lambda/H directly, we multiply the Schrödinger-like
# equation by H^2 and solve:
#   H_op psi = 4*pi^2 * (Lambda/H) psi
scale1 = np.sqrt(lambda_val) * Omega_val**2
scale2 = np.sqrt(lambda_val) * OmegaP_val**2

# ----------------------------------------------------------------------
# 3) 2D grid
# ----------------------------------------------------------------------
N1 = 151
N2 = 151

L1 = 4.0
L2 = 4.0

phi1_vals = np.linspace(-L1, L1, N1)
phi2_vals = np.linspace(-L2, L2, N2)

dphi1 = phi1_vals[1] - phi1_vals[0]
dphi2 = phi2_vals[1] - phi2_vals[0]

PHI1, PHI2 = np.meshgrid(phi1_vals, phi2_vals, indexing='ij')

# ----------------------------------------------------------------------
# 4) compute v and W_rescaled on the grid
# ----------------------------------------------------------------------
v_vals = v_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)

v1_vals  = v1_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)
v2_vals  = v2_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)
v11_vals = v11_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)
v22_vals = v22_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)

W_vals = 0.5 * (
    scale1 * (v1_vals**2 - v11_vals)
    +
    scale2 * (v2_vals**2 - v22_vals)
)

# ----------------------------------------------------------------------
# 5) find local minima of v on the grid
#    (simple 8-neighbour test)
# ----------------------------------------------------------------------
local_min_mask = np.ones_like(v_vals, dtype=bool)

for i_shift in [-1, 0, 1]:
    for j_shift in [-1, 0, 1]:
        if i_shift == 0 and j_shift == 0:
            continue
        shifted = np.roll(np.roll(v_vals, i_shift, axis=0), j_shift, axis=1)
        local_min_mask &= (v_vals <= shifted)

# remove boundary points from minima search
local_min_mask[0, :] = False
local_min_mask[-1, :] = False
local_min_mask[:, 0] = False
local_min_mask[:, -1] = False

min_indices = np.argwhere(local_min_mask)

if len(min_indices) < 2:
    raise RuntimeError("Could not find at least two local minima on this grid.")

min_points = []
for idx in min_indices:
    i, j = idx
    min_points.append((phi1_vals[i], phi2_vals[j], v_vals[i, j]))

min_points.sort(key=lambda x: x[2])  # sorted by v value

true_phi1, true_phi2, true_v = min_points[0]
false_phi1, false_phi2, false_v = min_points[1]

# ----------------------------------------------------------------------
# 6) simple straight-line path from false vacuum to true vacuum
# ----------------------------------------------------------------------
num_path = 300
s_vals = np.linspace(0.0, 1.0, num_path)

phi1_path = false_phi1 + (true_phi1 - false_phi1) * s_vals
phi2_path = false_phi2 + (true_phi2 - false_phi2) * s_vals

# ----------------------------------------------------------------------
# 7) plot v(phi1, phi2) with transition path
# ----------------------------------------------------------------------
fig, ax = plt.subplots(figsize=(8, 6))

cont = ax.contourf(PHI1, PHI2, v_vals, levels=40)
cbar = plt.colorbar(cont, ax=ax)
cbar.set_label(r'$v(\hat{\phi}_1,\hat{\phi}_2)$')

# straight-line path from false vacuum to true vacuum
ax.plot(phi1_path, phi2_path, 'w--', lw=2.2, label='Transition Path')

# mark start and end points
ax.scatter(false_phi1, false_phi2, color='red', s=70, zorder=5, label='False Vacuum')
ax.scatter(true_phi1,  true_phi2,  color='green', s=70, zorder=5, label='True Vacuum')

# arrow showing direction
mid = len(phi1_path) // 2
ax.annotate(
    '',
    xy=(phi1_path[mid + 8], phi2_path[mid + 8]),
    xytext=(phi1_path[mid - 8], phi2_path[mid - 8]),
    arrowprops=dict(arrowstyle='->', color='white', lw=2)
)

ax.set_xlabel(r'$\hat{\phi}_1$')
ax.set_ylabel(r'$\hat{\phi}_2$')
#ax.set_title(r'Dimensionless potential $v(\hat{\phi}_1,\hat{\phi}_2)$')

ax.legend(
    loc='upper right',
    frameon=True,
    facecolor='white',
    edgecolor='black',
    fontsize=9
)

info_text = (
    rf"$v_{{TV}} = {true_v:.4f}$" "\n"
    rf"$v_{{FV}} = {false_v:.4f}$"
)

ax.text(
    0.02, 0.97,
    info_text,
    transform=ax.transAxes,
    fontsize=10,
    verticalalignment='top',
    horizontalalignment='left',
    bbox=dict(boxstyle='round', facecolor='white', alpha=0.9, edgecolor='black')
)

#ax.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()

# ----------------------------------------------------------------------
# OPTIONAL plot of W_rescaled
# ----------------------------------------------------------------------
if SHOW_W:
    figW, axW = plt.subplots(figsize=(8, 6))

    contW = axW.contourf(PHI1, PHI2, W_vals, levels=40)
    cbarW = plt.colorbar(contW, ax=axW)
    cbarW.set_label(r'$W_{\mathrm{rescaled}}(\hat{\phi}_1,\hat{\phi}_2)$')

    axW.set_xlabel(r'$\hat{\phi}_1$')
    axW.set_ylabel(r'$\hat{\phi}_2$')
    axW.set_title(r'Rescaled effective potential $W_{\mathrm{rescaled}}$')
    axW.grid(True, alpha=0.3)

    plt.tight_layout()

    plt.show()

# ----------------------------------------------------------------------
# 8) build 2D Hamiltonian and solve spectrum
# ----------------------------------------------------------------------
# 1D Laplacians
main_diag_1 = np.full(N1, -2.0)
off_diag_1  = np.ones(N1 - 1)
lap1 = diags([off_diag_1, main_diag_1, off_diag_1], [-1, 0, 1], format='csr') / dphi1**2

main_diag_2 = np.full(N2, -2.0)
off_diag_2  = np.ones(N2 - 1)
lap2 = diags([off_diag_2, main_diag_2, off_diag_2], [-1, 0, 1], format='csr') / dphi2**2

I1 = eye(N1, format='csr')
I2 = eye(N2, format='csr')

# Kinetic term:
# T = -1/2 [ scale1 * d^2/dphi1^2 + scale2 * d^2/dphi2^2 ]
T1 = -0.5 * scale1 * kron(lap1, I2, format='csr')
T2 = -0.5 * scale2 * kron(I1, lap2, format='csr')
T = T1 + T2

# Potential term
V_op = diags(W_vals.reshape(N1 * N2), 0, format='csr')

# Full Hamiltonian:
# H_op psi = 4*pi^2 * (Lambda/H) psi
H_op = T + V_op

# Solve for lowest eigenvalues
num_eigs = 6
E_vals, _ = eigsh(H_op, k=num_eigs, sigma=0.0, which='LM')
E_vals = np.sort(E_vals)

Lambda_over_H = E_vals / (4 * np.pi**2)

print("\n n |   E_n (rescaled)   |   Lambda_n / H")
print("-" * 45)
for n in range(num_eigs):
    print(f"{n:1d} | {E_vals[n]: .6e} | {Lambda_over_H[n]: .6e}")
```

```python
# ----------------------------------------------------------------------
# 8) simple transition path from false vacuum to true vacuum
#    (straight-line interpolation in field space)
# ----------------------------------------------------------------------
# Note:
# This does NOT compute the true most-likely tunnelling path.
# It simply shows how phi1 and phi2 change along the direct line
# connecting the false and true vacua in field space.

num_path = 300
s_vals = np.linspace(0.0, 1.0, num_path)

phi1_path = false_phi1 + (true_phi1 - false_phi1) * s_vals
phi2_path = false_phi2 + (true_phi2 - false_phi2) * s_vals

v_path = v_func(phi1_path, phi2_path, alpha_val, beta_val, delta_val, gamma_val, xi_val)

# Plot how the fields change along the transition parameter s
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(s_vals, phi1_path, lw=2, label=r'$\hat{\phi}_1$')
ax.plot(s_vals, phi2_path, lw=2, label=r'$\hat{\phi}_2$')

ax.set_xlabel(r'Transition parameter $s$')
ax.set_ylabel(r'Field value')
ax.set_title(r'Change of $\hat{\phi}_1$ and $\hat{\phi}_2$ from FV to TV')
ax.grid(True, alpha=0.3)
ax.legend()

plt.tight_layout()
plt.show()

# Plot the potential along the same path
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(s_vals, v_path, lw=2)
ax.scatter(0, v_path[0], color='red', s=60, zorder=5, label='False Vacuum')
ax.scatter(1, v_path[-1], color='green', s=60, zorder=5, label='True Vacuum')

ax.set_xlabel(r'$s$')
ax.set_ylabel(r'$v(\hat{\phi}_1,\hat{\phi}_2)$ Along Path')
#ax.set_title(r'Potential along the straight-line transition path')
ax.grid(True, alpha=0.3)
ax.legend()

plt.tight_layout()

plt.show()

print("\nFalse vacuum coordinates:")
print(f"phi1_FV = {false_phi1:.6f}, phi2_FV = {false_phi2:.6f}")

print("\nTrue vacuum coordinates:")
print(f"phi1_TV = {true_phi1:.6f}, phi2_TV = {true_phi2:.6f}")
```

```python
# ----------------------------------------------------------------------
# 9) parameter scans with one parameter varied at a time
#    fixed values:
#    alpha = 0.1
#    beta  = 1.8
#    delta = 0.1
#    gamma = 1.8
#    xi    = 0.2
#    lambda= 0.05
# ----------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import diags, eye, kron
from scipy.sparse.linalg import eigsh

# -------------------------------------------------
# fixed parameters for scans
# -------------------------------------------------
alpha_fixed  = 0.1
beta_fixed   = 1.8
delta_fixed  = 0.1
gamma_fixed  = 1.8
xi_fixed     = 0.2
lambda_fixed = 0.05

# -------------------------------------------------
# scan grid
# -------------------------------------------------
N1_scan = 81
N2_scan = 81
L1_scan = 4.0
L2_scan = 4.0

phi1_scan = np.linspace(-L1_scan, L1_scan, N1_scan)
phi2_scan = np.linspace(-L2_scan, L2_scan, N2_scan)

dphi1_scan = phi1_scan[1] - phi1_scan[0]
dphi2_scan = phi2_scan[1] - phi2_scan[0]

PHI1_scan, PHI2_scan = np.meshgrid(phi1_scan, phi2_scan, indexing='ij')

# 1D Laplacians
main_diag_1 = np.full(N1_scan, -2.0)
off_diag_1  = np.ones(N1_scan - 1)
lap1_scan = diags([off_diag_1, main_diag_1, off_diag_1], [-1, 0, 1], format='csr') / dphi1_scan**2

main_diag_2 = np.full(N2_scan, -2.0)
off_diag_2  = np.ones(N2_scan - 1)
lap2_scan = diags([off_diag_2, main_diag_2, off_diag_2], [-1, 0, 1], format='csr') / dphi2_scan**2

I1_scan = eye(N1_scan, format='csr')
I2_scan = eye(N2_scan, format='csr')

# -------------------------------------------------
# helper function: compute Lambda_1/H and Lambda_2/H
# -------------------------------------------------
def compute_lambda12_over_H(alpha_val, beta_val, delta_val, gamma_val, xi_val, lambda_val,
                            k_eigs=6, zero_tol=1e-10):
    if beta_val <= 0 or gamma_val <= 0 or lambda_val <= 0:
        return np.nan, np.nan

    Omega_val  = 1 + alpha_val + np.sqrt(beta_val)
    OmegaP_val = 1 + delta_val + np.sqrt(gamma_val)

    scale1 = np.sqrt(lambda_val) * Omega_val**2
    scale2 = np.sqrt(lambda_val) * OmegaP_val**2

    v1_vals  = v1_func(PHI1_scan, PHI2_scan, alpha_val, beta_val, delta_val, gamma_val, xi_val)
    v2_vals  = v2_func(PHI1_scan, PHI2_scan, alpha_val, beta_val, delta_val, gamma_val, xi_val)
    v11_vals = v11_func(PHI1_scan, PHI2_scan, alpha_val, beta_val, delta_val, gamma_val, xi_val)
    v22_vals = v22_func(PHI1_scan, PHI2_scan, alpha_val, beta_val, delta_val, gamma_val, xi_val)

    W_vals = 0.5 * (
        scale1 * (v1_vals**2 - v11_vals)
        +
        scale2 * (v2_vals**2 - v22_vals)
    )

    T1 = -0.5 * scale1 * kron(lap1_scan, I2_scan, format='csr')
    T2 = -0.5 * scale2 * kron(I1_scan, lap2_scan, format='csr')
    T = T1 + T2

    V_op = diags(W_vals.reshape(N1_scan * N2_scan), 0, format='csr')
    H_op = T + V_op

    E_vals, _ = eigsh(H_op, k=k_eigs, sigma=0.0, which='LM')
    E_vals = np.sort(E_vals)

    Lambda_over_H = E_vals / (4 * np.pi**2)
    positive_modes = Lambda_over_H[Lambda_over_H > zero_tol]

    if len(positive_modes) < 2:
        return np.nan, np.nan

    return positive_modes[0], positive_modes[1]

# -------------------------------------------------
# helper function: make one plot
# -------------------------------------------------
def make_scan_plot(xvals, label, param_name):
    Lambda1 = []
    Lambda2 = []

    for val in xvals:
        a = alpha_fixed
        b = beta_fixed
        d = delta_fixed
        g = gamma_fixed
        x = xi_fixed
        lam = lambda_fixed

        if param_name == 'alpha':
            a = val
        elif param_name == 'beta':
            b = val
        elif param_name == 'delta':
            d = val
        elif param_name == 'gamma':
            g = val
        elif param_name == 'xi':
            x = val
        elif param_name == 'lambda':
            lam = val

        lam1, lam2 = compute_lambda12_over_H(a, b, d, g, x, lam)
        Lambda1.append(lam1)
        Lambda2.append(lam2)

    Lambda1 = np.array(Lambda1)
    Lambda2 = np.array(Lambda2)

    mask = np.isfinite(Lambda1) & np.isfinite(Lambda2) & (Lambda1 > 0) & (Lambda2 > 0)

    plt.figure(figsize=(7, 5))

    plt.plot(
        xvals[mask],
        Lambda1[mask],
        lw=2,
        color='tab:blue',
        label=r'$\Lambda_1/H$'
    )

    plt.plot(
        xvals[mask],
        Lambda2[mask],
        lw=2,
        color='gold',
        alpha=0.95,
        label=r'$\Lambda_2/H$'
    )

    plt.xlabel(label)
    plt.ylabel(r'$\Lambda_n/H$')
    plt.yscale('log')
    plt.grid(True)
    plt.legend()

    fixed_lines = []

    if param_name != 'alpha':
        fixed_lines.append(rf"$\alpha = {alpha_fixed}$")
    if param_name != 'beta':
        fixed_lines.append(rf"$\beta = {beta_fixed}$")
    if param_name != 'delta':
        fixed_lines.append(rf"$\delta = {delta_fixed}$")
    if param_name != 'gamma':
        fixed_lines.append(rf"$\gamma = {gamma_fixed}$")
    if param_name != 'xi':
        fixed_lines.append(rf"$\xi = {xi_fixed}$")
    if param_name != 'lambda':
        fixed_lines.append(rf"$\lambda = {lambda_fixed}$")

    fixed_text = "\n".join(fixed_lines)

    plt.text(
        0.03, 0.97,
        fixed_text,
        transform=plt.gca().transAxes,
        fontsize=10,
        verticalalignment='top',
        horizontalalignment='left',
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.9, edgecolor='black')
    )

    plt.tight_layout()
    #plt.savefig(f"Lambda_scan_{param_name}.png", dpi=300, bbox_inches='tight')
    plt.show()

# -------------------------------------------------
# produce the 6 plots with your requested ranges
# -------------------------------------------------
alpha_vals  = np.linspace(0.2, 1.2, 120)
beta_vals   = np.linspace(0.5, 1.5, 120)
delta_vals  = np.linspace(0.2, 1.2, 120)
gamma_vals  = np.linspace(0.5, 1.5, 120)
xi_vals     = np.linspace(0.2, 1.0, 120)
lambda_vals = np.linspace(0.2, 1.0, 120)

make_scan_plot(alpha_vals,  r'$\alpha$',  'alpha')
make_scan_plot(beta_vals,   r'$\beta$',   'beta')
make_scan_plot(delta_vals,  r'$\delta$',  'delta')
make_scan_plot(gamma_vals,  r'$\gamma$',  'gamma')
make_scan_plot(xi_vals,     r'$\xi$',     'xi')
make_scan_plot(lambda_vals, r'$\lambda$', 'lambda')
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from scipy.sparse import diags, eye, kron
from scipy.sparse.linalg import eigsh
from scipy.optimize import differential_evolution

# =========================
# ONE SIMPLE TOGGLE
# =========================
SHOW_W = False   # True: show W_rescaled contour, False: hide it

# ----------------------------------------------------------------------
# 1) symbolic definitions
# ----------------------------------------------------------------------
phi1, phi2 = sp.symbols('phi1 phi2', real=True)
alpha, beta, delta, gamma, xi = sp.symbols('alpha beta delta gamma xi', real=True)

Omega  = 1 + alpha + sp.sqrt(beta)
OmegaP = 1 + delta + sp.sqrt(gamma)

v = (
    (sp.pi**2 / (3 * Omega)) * (
        4*alpha*phi1
        - 2*(beta/Omega)*phi1**2
        + (4/Omega**3)*phi1**4
    )
    +
    (sp.pi**2 / (3 * OmegaP)) * (
        4*delta*phi2
        - 2*(gamma/OmegaP)*phi2**2
        + (4/OmegaP**3)*phi2**4
    )
    +
    (sp.pi**2 / (3 * Omega * OmegaP)) * xi * phi1**2 * phi2**2
)

v_phi1  = sp.diff(v, phi1)
v_phi2  = sp.diff(v, phi2)
v_phi11 = sp.diff(v_phi1, phi1)
v_phi22 = sp.diff(v_phi2, phi2)

v_func   = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v, 'numpy')
v1_func  = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi1, 'numpy')
v2_func  = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi2, 'numpy')
v11_func = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi11, 'numpy')
v22_func = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi22, 'numpy')

# ----------------------------------------------------------------------
# 2) target vacuum/saddle values
# ----------------------------------------------------------------------
target_true_v   = -5.9891
target_false_v  = -2.0929
target_barrier  = 0.0

# lambda does not affect v; keep it fixed for the spectrum
lambda_val = 0.05

# ----------------------------------------------------------------------
# 3) coarse grid for parameter search
# ----------------------------------------------------------------------
N_search = 121
L1_search = 4.0
L2_search = 4.0

phi1_search = np.linspace(-L1_search, L1_search, N_search)
phi2_search = np.linspace(-L2_search, L2_search, N_search)
PHI1s, PHI2s = np.meshgrid(phi1_search, phi2_search, indexing='ij')

# ----------------------------------------------------------------------
# 4) helper: find two minima + barrier along straight-line path
# ----------------------------------------------------------------------
def extract_vacua_and_barrier(alpha_val, beta_val, delta_val, gamma_val, xi_val, num_path=400):
    if beta_val <= 0 or gamma_val <= 0:
        return None

    v_vals = v_func(PHI1s, PHI2s, alpha_val, beta_val, delta_val, gamma_val, xi_val)

    # local minima search
    local_min_mask = np.ones_like(v_vals, dtype=bool)

    for i_shift in [-1, 0, 1]:
        for j_shift in [-1, 0, 1]:
            if i_shift == 0 and j_shift == 0:
                continue
            shifted = np.roll(np.roll(v_vals, i_shift, axis=0), j_shift, axis=1)
            local_min_mask &= (v_vals <= shifted)

    # remove boundaries
    local_min_mask[0, :]  = False
    local_min_mask[-1, :] = False
    local_min_mask[:, 0]  = False
    local_min_mask[:, -1] = False

    min_indices = np.argwhere(local_min_mask)
    if len(min_indices) < 2:
        return None

    min_points = []
    for idx in min_indices:
        i, j = idx
        min_points.append((phi1_search[i], phi2_search[j], v_vals[i, j]))

    min_points.sort(key=lambda x: x[2])

    true_phi1, true_phi2, true_v = min_points[0]
    false_phi1, false_phi2, false_v = min_points[1]

    # straight-line path
    s_vals = np.linspace(0.0, 1.0, num_path)
    phi1_path = false_phi1 + (true_phi1 - false_phi1) * s_vals
    phi2_path = false_phi2 + (true_phi2 - false_phi2) * s_vals

    v_path = v_func(phi1_path, phi2_path, alpha_val, beta_val, delta_val, gamma_val, xi_val)
    barrier_v = np.max(v_path)

    return {
        "true_phi1": true_phi1,
        "true_phi2": true_phi2,
        "true_v": true_v,
        "false_phi1": false_phi1,
        "false_phi2": false_phi2,
        "false_v": false_v,
        "barrier_v": barrier_v,
    }

# ----------------------------------------------------------------------
# 5) objective function for parameter tuning
# ----------------------------------------------------------------------
def objective(params):
    alpha_val, beta_val, delta_val, gamma_val, xi_val = params

    data = extract_vacua_and_barrier(alpha_val, beta_val, delta_val, gamma_val, xi_val)
    if data is None:
        return 1e6

    true_v = data["true_v"]
    false_v = data["false_v"]
    barrier_v = data["barrier_v"]

    # enforce TV deeper than FV
    if not (true_v < false_v < barrier_v):
        return 1e6

    # squared error target matching
    loss = (
        (true_v   - target_true_v )**2 +
        (false_v  - target_false_v)**2 +
        (barrier_v - target_barrier)**2
    )

    # optional small penalty if barrier is not sufficiently above false vacuum
    if barrier_v <= false_v:
        loss += 1e5

    return loss

# ----------------------------------------------------------------------
# 6) parameter search
# ----------------------------------------------------------------------
bounds = [
    (0.0, 1.5),   # alpha
    (0.5, 2.5),   # beta
    (0.0, 1.5),   # delta
    (0.5, 2.5),   # gamma
    (0.0, 1.0),   # xi
]

result = differential_evolution(
    objective,
    bounds=bounds,
    maxiter=40,
    popsize=12,
    tol=1e-3,
    polish=True,
    seed=42
)

alpha_val, beta_val, delta_val, gamma_val, xi_val = result.x

best_data = extract_vacua_and_barrier(alpha_val, beta_val, delta_val, gamma_val, xi_val)

print("\nBest-fit parameters found:")
print(f"alpha  = {alpha_val:.6f}")
print(f"beta   = {beta_val:.6f}")
print(f"delta  = {delta_val:.6f}")
print(f"gamma  = {gamma_val:.6f}")
print(f"xi     = {xi_val:.6f}")
print(f"lambda = {lambda_val:.6f}")

print("\nMatched potential values:")
print(f"true vacuum   = {best_data['true_v']:.6f}   (target {target_true_v})")
print(f"false vacuum  = {best_data['false_v']:.6f}   (target {target_false_v})")
print(f"barrier top   = {best_data['barrier_v']:.6f}   (target {target_barrier})")

# ----------------------------------------------------------------------
# 7) parameters for the full calculation
# ----------------------------------------------------------------------
Omega_val  = 1 + alpha_val + np.sqrt(beta_val)
OmegaP_val = 1 + delta_val + np.sqrt(gamma_val)

scale1 = np.sqrt(lambda_val) * Omega_val**2
scale2 = np.sqrt(lambda_val) * OmegaP_val**2

# ----------------------------------------------------------------------
# 8) full 2D grid
# ----------------------------------------------------------------------
N1 = 151
N2 = 151

L1 = 4.0
L2 = 4.0

phi1_vals = np.linspace(-L1, L1, N1)
phi2_vals = np.linspace(-L2, L2, N2)

dphi1 = phi1_vals[1] - phi1_vals[0]
dphi2 = phi2_vals[1] - phi2_vals[0]

PHI1, PHI2 = np.meshgrid(phi1_vals, phi2_vals, indexing='ij')

# ----------------------------------------------------------------------
# 9) compute v and W_rescaled on the grid
# ----------------------------------------------------------------------
v_vals = v_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)

v1_vals  = v1_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)
v2_vals  = v2_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)
v11_vals = v11_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)
v22_vals = v22_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)

W_vals = 0.5 * (
    scale1 * (v1_vals**2 - v11_vals)
    +
    scale2 * (v2_vals**2 - v22_vals)
)

# ----------------------------------------------------------------------
# 10) find local minima of v on the full grid
# ----------------------------------------------------------------------
local_min_mask = np.ones_like(v_vals, dtype=bool)

for i_shift in [-1, 0, 1]:
    for j_shift in [-1, 0, 1]:
        if i_shift == 0 and j_shift == 0:
            continue
        shifted = np.roll(np.roll(v_vals, i_shift, axis=0), j_shift, axis=1)
        local_min_mask &= (v_vals <= shifted)

local_min_mask[0, :] = False
local_min_mask[-1, :] = False
local_min_mask[:, 0] = False
local_min_mask[:, -1] = False

min_indices = np.argwhere(local_min_mask)
if len(min_indices) < 2:
    raise RuntimeError("Could not find two local minima on the full grid.")

min_points = []
for idx in min_indices:
    i, j = idx
    min_points.append((phi1_vals[i], phi2_vals[j], v_vals[i, j]))

min_points.sort(key=lambda x: x[2])

true_phi1, true_phi2, true_v = min_points[0]
false_phi1, false_phi2, false_v = min_points[1]

# ----------------------------------------------------------------------
# 11) straight-line path and barrier along it
# ----------------------------------------------------------------------
num_path = 400
s_vals = np.linspace(0.0, 1.0, num_path)

phi1_path = false_phi1 + (true_phi1 - false_phi1) * s_vals
phi2_path = false_phi2 + (true_phi2 - false_phi2) * s_vals

v_path = v_func(phi1_path, phi2_path, alpha_val, beta_val, delta_val, gamma_val, xi_val)
barrier_idx = np.argmax(v_path)
barrier_v = v_path[barrier_idx]
barrier_phi1 = phi1_path[barrier_idx]
barrier_phi2 = phi2_path[barrier_idx]

# ----------------------------------------------------------------------
# 12) plot v(phi1, phi2) with transition path
# ----------------------------------------------------------------------
fig, ax = plt.subplots(figsize=(8, 6))

cont = ax.contourf(PHI1, PHI2, v_vals, levels=40)
cbar = plt.colorbar(cont, ax=ax)
cbar.set_label(r'$v(\hat{\phi}_1,\hat{\phi}_2)$')

ax.plot(phi1_path, phi2_path, 'w--', lw=2.2, label='Transition path')

ax.scatter(false_phi1, false_phi2, color='red', s=70, zorder=5, label='False Vacuum')
ax.scatter(true_phi1,  true_phi2,  color='green', s=70, zorder=5, label='True Vacuum')
ax.scatter(barrier_phi1, barrier_phi2, color='cyan', s=70, zorder=6, label='Barrier Top')

mid = len(phi1_path) // 2
ax.annotate(
    '',
    xy=(phi1_path[mid + 8], phi2_path[mid + 8]),
    xytext=(phi1_path[mid - 8], phi2_path[mid - 8]),
    arrowprops=dict(arrowstyle='->', color='white', lw=2)
)

ax.set_xlabel(r'$\hat{\phi}_1$')
ax.set_ylabel(r'$\hat{\phi}_2$')
#ax.set_title(r'Dimensionless potential $v(\hat{\phi}_1,\hat{\phi}_2)$')

ax.legend(
    loc='upper right',
    frameon=True,
    facecolor='white',
    edgecolor='black',
    fontsize=9
)

info_text = (
    rf"$v_{{TV}} = {true_v:.4f}$" "\n"
    rf"$v_{{FV}} = {false_v:.4f}$" 
)

ax.text(
    0.02, 0.97,
    info_text,
    transform=ax.transAxes,
    fontsize=10,
    verticalalignment='top',
    horizontalalignment='left',
    bbox=dict(boxstyle='round', facecolor='white', alpha=0.9, edgecolor='black')
)

ax.grid(True, alpha=0.3)
plt.tight_layout()

plt.show()

# ----------------------------------------------------------------------
# OPTIONAL plot of W_rescaled
# ----------------------------------------------------------------------
if SHOW_W:
    figW, axW = plt.subplots(figsize=(8, 6))

    contW = axW.contourf(PHI1, PHI2, W_vals, levels=40)
    cbarW = plt.colorbar(contW, ax=axW)
    cbarW.set_label(r'$W_{\mathrm{rescaled}}(\hat{\phi}_1,\hat{\phi}_2)$')

    axW.set_xlabel(r'$\hat{\phi}_1$')
    axW.set_ylabel(r'$\hat{\phi}_2$')
    axW.set_title(r'Rescaled effective potential $W_{\mathrm{rescaled}}$')
    axW.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# ----------------------------------------------------------------------
# 13) build 2D Hamiltonian and solve spectrum
# ----------------------------------------------------------------------
main_diag_1 = np.full(N1, -2.0)
off_diag_1  = np.ones(N1 - 1)
lap1 = diags([off_diag_1, main_diag_1, off_diag_1], [-1, 0, 1], format='csr') / dphi1**2

main_diag_2 = np.full(N2, -2.0)
off_diag_2  = np.ones(N2 - 1)
lap2 = diags([off_diag_2, main_diag_2, off_diag_2], [-1, 0, 1], format='csr') / dphi2**2

I1 = eye(N1, format='csr')
I2 = eye(N2, format='csr')

T1 = -0.5 * scale1 * kron(lap1, I2, format='csr')
T2 = -0.5 * scale2 * kron(I1, lap2, format='csr')
T = T1 + T2

V_op = diags(W_vals.reshape(N1 * N2), 0, format='csr')
H_op = T + V_op

num_eigs = 6
E_vals, _ = eigsh(H_op, k=num_eigs, sigma=0.0, which='LM')
E_vals = np.sort(E_vals)

Lambda_over_H = E_vals / (4 * np.pi**2)

print("\n n |   E_n (rescaled)   |   Lambda_n / H")
print("-" * 45)
for n in range(num_eigs):
    print(f"{n:1d} | {E_vals[n]: .6e} | {Lambda_over_H[n]: .6e}")
```

```python
# ----------------------------------------------------------------------
# 8) simple transition path from false vacuum to true vacuum
#    (straight-line interpolation in field space)
# ----------------------------------------------------------------------
# Note:
# This does NOT compute the true most-likely tunnelling path.
# It simply shows how phi1 and phi2 change along the direct line
# connecting the false and true vacua in field space.

num_path = 300
s_vals = np.linspace(0.0, 1.0, num_path)

phi1_path = false_phi1 + (true_phi1 - false_phi1) * s_vals
phi2_path = false_phi2 + (true_phi2 - false_phi2) * s_vals

v_path = v_func(phi1_path, phi2_path, alpha_val, beta_val, delta_val, gamma_val, xi_val)

# Plot how the fields change along the transition parameter s
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(s_vals, phi1_path, lw=2, label=r'$\hat{\phi}_1$')
ax.plot(s_vals, phi2_path, lw=2, label=r'$\hat{\phi}_2$')

ax.set_xlabel(r'Transition parameter $s$')
ax.set_ylabel(r'Field value')
ax.set_title(r'Change of $\hat{\phi}_1$ and $\hat{\phi}_2$ from FV to TV')
ax.grid(True, alpha=0.3)
ax.legend()

plt.tight_layout()
plt.show()

# Plot the potential along the same path
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(s_vals, v_path, lw=2)
ax.scatter(0, v_path[0], color='red', s=60, zorder=5, label='False Vacuum')
ax.scatter(1, v_path[-1], color='green', s=60, zorder=5, label='True Vacuum')

ax.set_xlabel(r'$s$')
ax.set_ylabel(r'$v(\hat{\phi}_1,\hat{\phi}_2)$ Along Path')
#ax.set_title(r'Potential along the straight-line transition path')
ax.grid(True, alpha=0.3)
ax.legend()

plt.tight_layout()

plt.show()

print("\nFalse vacuum coordinates:")
print(f"phi1_FV = {false_phi1:.6f}, phi2_FV = {false_phi2:.6f}")

print("\nTrue vacuum coordinates:")
print(f"phi1_TV = {true_phi1:.6f}, phi2_TV = {true_phi2:.6f}")
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

from scipy.sparse import diags, eye, kron
from scipy.sparse.linalg import eigsh
from scipy.optimize import differential_evolution

# =========================
# ONE SIMPLE TOGGLE
# =========================
SHOW_W = False   # True: show W_rescaled contour, False: hide it

# ----------------------------------------------------------------------
# 1) symbolic definitions
# ----------------------------------------------------------------------
phi1, phi2 = sp.symbols('phi1 phi2', real=True)
alpha, beta, delta, gamma, xi = sp.symbols('alpha beta delta gamma xi', real=True)

Omega  = 1 + alpha + sp.sqrt(beta)
OmegaP = 1 + delta + sp.sqrt(gamma)

v = (
    (sp.pi**2 / (3 * Omega)) * (
        4*alpha*phi1
        - 2*(beta/Omega)*phi1**2
        + (4/Omega**3)*phi1**4
    )
    +
    (sp.pi**2 / (3 * OmegaP)) * (
        4*delta*phi2
        - 2*(gamma/OmegaP)*phi2**2
        + (4/OmegaP**3)*phi2**4
    )
    +
    (sp.pi**2 / (3 * Omega * OmegaP)) * xi * phi1**2 * phi2**2
)

v_phi1  = sp.diff(v, phi1)
v_phi2  = sp.diff(v, phi2)
v_phi11 = sp.diff(v_phi1, phi1)
v_phi22 = sp.diff(v_phi2, phi2)

v_func   = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v, 'numpy')
v1_func  = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi1, 'numpy')
v2_func  = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi2, 'numpy')
v11_func = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi11, 'numpy')
v22_func = sp.lambdify((phi1, phi2, alpha, beta, delta, gamma, xi), v_phi22, 'numpy')

# ----------------------------------------------------------------------
# 2) fit parameters to target minima (approximate best fit)
# ----------------------------------------------------------------------
# target minima: (phi1, phi2, value)
target_minima = [
    (-1.03307490, -1.01070994, -2.155440781009),
    (-1.06070107,  0.63165234, -1.331421005870),
    ( 0.82362678, -1.04481603, -0.838992985666),
    ( 0.86314065,  0.72277428,  0.029424066421),
]

# coarse grid for fitting
N_fit = 121
L1_fit = 4.0
L2_fit = 4.0
phi1_fit = np.linspace(-L1_fit, L1_fit, N_fit)
phi2_fit = np.linspace(-L2_fit, L2_fit, N_fit)
PHI1_fit, PHI2_fit = np.meshgrid(phi1_fit, phi2_fit, indexing='ij')

def find_minima_for_params(alpha_val, beta_val, delta_val, gamma_val, xi_val):
    if beta_val <= 0 or gamma_val <= 0:
        return None

    v_vals_fit = v_func(PHI1_fit, PHI2_fit, alpha_val, beta_val, delta_val, gamma_val, xi_val)

    local_min_mask = np.ones_like(v_vals_fit, dtype=bool)

    for i_shift in [-1, 0, 1]:
        for j_shift in [-1, 0, 1]:
            if i_shift == 0 and j_shift == 0:
                continue
            shifted = np.roll(np.roll(v_vals_fit, i_shift, axis=0), j_shift, axis=1)
            local_min_mask &= (v_vals_fit <= shifted)

    local_min_mask[0, :] = False
    local_min_mask[-1, :] = False
    local_min_mask[:, 0] = False
    local_min_mask[:, -1] = False

    min_indices = np.argwhere(local_min_mask)
    if len(min_indices) < 4:
        return None

    min_points = []
    for idx in min_indices:
        i, j = idx
        min_points.append((phi1_fit[i], phi2_fit[j], v_vals_fit[i, j]))

    min_points.sort(key=lambda x: x[2])
    return min_points[:4]

def objective(params):
    alpha_val, beta_val, delta_val, gamma_val, xi_val = params

    mins = find_minima_for_params(alpha_val, beta_val, delta_val, gamma_val, xi_val)
    if mins is None:
        return 1e8

    loss = 0.0
    for (x_t, y_t, v_t), (x_m, y_m, v_m) in zip(target_minima, mins):
        loss += 30.0 * (x_m - x_t)**2
        loss += 30.0 * (y_m - y_t)**2
        loss += 10.0 * (v_m - v_t)**2

    return loss

bounds = [
    (0.0, 1.5),   # alpha
    (0.5, 3.0),   # beta
    (0.0, 1.5),   # delta
    (0.5, 3.0),   # gamma
    (0.0, 1.0),   # xi
]

result = differential_evolution(
    objective,
    bounds=bounds,
    maxiter=50,
    popsize=15,
    tol=1e-4,
    polish=True,
    seed=42
)

alpha_val, beta_val, delta_val, gamma_val, xi_val = result.x

print("\nBest-fit parameters:")
print(f"alpha_val = {alpha_val:.8f}")
print(f"beta_val  = {beta_val:.8f}")
print(f"delta_val = {delta_val:.8f}")
print(f"gamma_val = {gamma_val:.8f}")
print(f"xi_val    = {xi_val:.8f}")

best_mins = find_minima_for_params(alpha_val, beta_val, delta_val, gamma_val, xi_val)

print("\nBest-fit minima on fitting grid (sorted by potential value):")
for k, (x, y, val) in enumerate(best_mins, start=1):
    print(f"  minimum {k}:")
    print(f"    coordinates = ({x: .8f}, {y: .8f})")
    print(f"    value       = {val: .12f}")

print("\nTarget minima:")
for k, (x, y, val) in enumerate(target_minima, start=1):
    print(f"  minimum {k}:")
    print(f"    coordinates = ({x: .8f}, {y: .8f})")
    print(f"    value       = {val: .12f}")

# ----------------------------------------------------------------------
# 3) remaining parameter
# ----------------------------------------------------------------------
lambda_val = 0.05

Omega_val  = 1 + alpha_val + np.sqrt(beta_val)
OmegaP_val = 1 + delta_val + np.sqrt(gamma_val)

# Since we want Lambda/H directly, we multiply the Schrödinger-like
# equation by H^2 and solve:
#   H_op psi = 4*pi^2 * (Lambda/H) psi
scale1 = np.sqrt(lambda_val) * Omega_val**2
scale2 = np.sqrt(lambda_val) * OmegaP_val**2

# ----------------------------------------------------------------------
# 4) 2D grid
# ----------------------------------------------------------------------
N1 = 151
N2 = 151

L1 = 4.0
L2 = 4.0

phi1_vals = np.linspace(-L1, L1, N1)
phi2_vals = np.linspace(-L2, L2, N2)

dphi1 = phi1_vals[1] - phi1_vals[0]
dphi2 = phi2_vals[1] - phi2_vals[0]

PHI1, PHI2 = np.meshgrid(phi1_vals, phi2_vals, indexing='ij')

# ----------------------------------------------------------------------
# 5) compute v and W_rescaled on the grid
# ----------------------------------------------------------------------
v_vals = v_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)

v1_vals  = v1_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)
v2_vals  = v2_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)
v11_vals = v11_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)
v22_vals = v22_func(PHI1, PHI2, alpha_val, beta_val, delta_val, gamma_val, xi_val)

W_vals = 0.5 * (
    scale1 * (v1_vals**2 - v11_vals)
    +
    scale2 * (v2_vals**2 - v22_vals)
)

# ----------------------------------------------------------------------
# 6) find local minima of v on the grid
#    (simple 8-neighbour test)
# ----------------------------------------------------------------------
local_min_mask = np.ones_like(v_vals, dtype=bool)

for i_shift in [-1, 0, 1]:
    for j_shift in [-1, 0, 1]:
        if i_shift == 0 and j_shift == 0:
            continue
        shifted = np.roll(np.roll(v_vals, i_shift, axis=0), j_shift, axis=1)
        local_min_mask &= (v_vals <= shifted)

# remove boundary points from minima search
local_min_mask[0, :] = False
local_min_mask[-1, :] = False
local_min_mask[:, 0] = False
local_min_mask[:, -1] = False

min_indices = np.argwhere(local_min_mask)

if len(min_indices) < 2:
    raise RuntimeError("Could not find at least two local minima on this grid.")

min_points = []
for idx in min_indices:
    i, j = idx
    min_points.append((phi1_vals[i], phi2_vals[j], v_vals[i, j]))

min_points.sort(key=lambda x: x[2])  # sorted by v value

print("\nList of minima on main grid (sorted by potential value):")
for k, (x, y, val) in enumerate(min_points[:10], start=1):
    print(f"  minimum {k}:")
    print(f"    coordinates = ({x: .8f}, {y: .8f})")
    print(f"    value       = {val: .12f}")

true_phi1, true_phi2, true_v = min_points[0]
false_phi1, false_phi2, false_v = min_points[1]

# ----------------------------------------------------------------------
# 7) simple straight-line path from false vacuum to true vacuum
# ----------------------------------------------------------------------
num_path = 300
s_vals = np.linspace(0.0, 1.0, num_path)

phi1_path = false_phi1 + (true_phi1 - false_phi1) * s_vals
phi2_path = false_phi2 + (true_phi2 - false_phi2) * s_vals

# ----------------------------------------------------------------------
# 8) plot v(phi1, phi2) with transition path
# ----------------------------------------------------------------------
fig, ax = plt.subplots(figsize=(8, 6))

cont = ax.contourf(PHI1, PHI2, v_vals, levels=40)
cbar = plt.colorbar(cont, ax=ax)
cbar.set_label(r'$v(\hat{\phi}_1,\hat{\phi}_2)$')

# straight-line path from false vacuum to true vacuum
ax.plot(phi1_path, phi2_path, 'w--', lw=2.2, label='Transition Path')

# mark start and end points
ax.scatter(false_phi1, false_phi2, color='red', s=70, zorder=5, label='False Vacuum')
ax.scatter(true_phi1,  true_phi2,  color='green', s=70, zorder=5, label='True Vacuum')

# arrow showing direction
mid = len(phi1_path) // 2
ax.annotate(
    '',
    xy=(phi1_path[mid + 8], phi2_path[mid + 8]),
    xytext=(phi1_path[mid - 8], phi2_path[mid - 8]),
    arrowprops=dict(arrowstyle='->', color='white', lw=2)
)

ax.set_xlabel(r'$\hat{\phi}_1$')
ax.set_ylabel(r'$\hat{\phi}_2$')

ax.legend(
    loc='upper right',
    frameon=True,
    facecolor='white',
    edgecolor='black',
    fontsize=9
)

info_text = (
    rf"$v_{{TV}} = {true_v:.4f}$" "\n"
    rf"$v_{{FV}} = {false_v:.4f}$"
)

ax.text(
    0.02, 0.97,
    info_text,
    transform=ax.transAxes,
    fontsize=10,
    verticalalignment='top',
    horizontalalignment='left',
    bbox=dict(boxstyle='round', facecolor='white', alpha=0.9, edgecolor='black')
)

plt.tight_layout()

plt.show()

# ----------------------------------------------------------------------
# OPTIONAL plot of W_rescaled
# ----------------------------------------------------------------------
if SHOW_W:
    figW, axW = plt.subplots(figsize=(8, 6))

    contW = axW.contourf(PHI1, PHI2, W_vals, levels=40)
    cbarW = plt.colorbar(contW, ax=axW)
    cbarW.set_label(r'$W_{\mathrm{rescaled}}(\hat{\phi}_1,\hat{\phi}_2)$')

    axW.set_xlabel(r'$\hat{\phi}_1$')
    axW.set_ylabel(r'$\hat{\phi}_2$')
    axW.set_title(r'Rescaled effective potential $W_{\mathrm{rescaled}}$')
    axW.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

# ----------------------------------------------------------------------
# 9) build 2D Hamiltonian and solve spectrum
# ----------------------------------------------------------------------
# 1D Laplacians
main_diag_1 = np.full(N1, -2.0)
off_diag_1  = np.ones(N1 - 1)
lap1 = diags([off_diag_1, main_diag_1, off_diag_1], [-1, 0, 1], format='csr') / dphi1**2

main_diag_2 = np.full(N2, -2.0)
off_diag_2  = np.ones(N2 - 1)
lap2 = diags([off_diag_2, main_diag_2, off_diag_2], [-1, 0, 1], format='csr') / dphi2**2

I1 = eye(N1, format='csr')
I2 = eye(N2, format='csr')

# Kinetic term:
# T = -1/2 [ scale1 * d^2/dphi1^2 + scale2 * d^2/dphi2^2 ]
T1 = -0.5 * scale1 * kron(lap1, I2, format='csr')
T2 = -0.5 * scale2 * kron(I1, lap2, format='csr')
T = T1 + T2

# Potential term
V_op = diags(W_vals.reshape(N1 * N2), 0, format='csr')

# Full Hamiltonian:
# H_op psi = 4*pi^2 * (Lambda/H) psi
H_op = T + V_op

# Solve for lowest eigenvalues
num_eigs = 6
E_vals, _ = eigsh(H_op, k=num_eigs, sigma=0.0, which='LM')
E_vals = np.sort(E_vals)

Lambda_over_H = E_vals / (4 * np.pi**2)

print("\n n |   E_n (rescaled)   |   Lambda_n / H")
print("-" * 45)
for n in range(num_eigs):
    print(f"{n:1d} | {E_vals[n]: .6e} | {Lambda_over_H[n]: .6e}")
```
